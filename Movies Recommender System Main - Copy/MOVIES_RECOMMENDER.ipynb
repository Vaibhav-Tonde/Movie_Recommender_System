```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on='title') #merges the content on title
```

```python
movies.head(1)
```

```python
#what we are adding to project
#geners
#id
#keywords
#title
#overview
#cast
#crew

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True) # drop the null data
```

```python
movies.duplicated().sum() #checks the duplicate  data.
```

```python
movies.iloc[0].genres # iloc() function enables us to select a particular cell of the dataset
```

```python
# to convert this data in specific list. ['Action','Adventure','Fantasy','Sci-Fi']
# ast.literal_eval converts the string list to list.
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
  movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fetch_diractor(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
        
```

```python
movies['crew'] = movies['crew'].apply(fetch_diractor)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies.head()
```

```python
#now we have to remove spaces because For. Ex Sam Worthington here Sam is other entity and Worthington is other so our system
#will not confuse with same name.
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
movies['tags'] =  movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id','title','overview','tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
# here's point how do we say these 5 movies are similar?
# ----> so we need to find the similarity between two movies.
# ---->we will use text-vectorization.
# we will use bag-of-words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags'],'overview').toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
# we will use "cosine distance" between movie vectors insted of "euclidian distance".
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']== movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    return
```

```python
recommend('The Dark Knight Rises')
```
